```python
from sklearn.base import BaseEstimator, TransformerMixin

square_ix = df.columns.get_loc("Общая")
living_ix = df.columns.get_loc("Жилая")
kitchen_ix = df.columns.get_loc("Кухня")

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_freeplace = True):
        self.add_freeplace = add_freeplace
    def fit(self, X, y=None):
        return self  
    def transform(self, X):
        if self.add_freeplace:
            freeplace = X.loc[:, square_ix] - X.loc[:, living_ix] - X.loc[:, kitchen_ix]
            return np.c_[X, freeplace]
        
```

```python
class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names]
```

```python
sns.distplot(df["цена"] , fit=norm);

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(df["цена"])
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

#Now plot the distribution
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

#Get also the QQ-plot
fig = plt.figure()
res = stats.probplot(df["цена"], plot=plt)
plt.show()
```

```python
#We use the numpy fuction log1p which  applies log(1+x) to all elements of the column
price_log = np.log1p(df["цена"])

#Check the new distribution 
sns.distplot(price_log , fit=norm);

# Get the fitted parameters used by the function
(mu, sigma) = norm.fit(price_log)
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

#Now plot the distribution
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='best')
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

#Get also the QQ-plot
fig = plt.figure()
res = stats.probplot(price_log,  plot=plt)
plt.show()
```
